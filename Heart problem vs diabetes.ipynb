```python
fig = plt.figure(figsize = (20,15))
ax = fig.gca()
data.hist(ax=ax)
plt.show()
```

```python
data[['diabetes', 'DEATH_EVENT']].describe()
```

```python
sns.countplot(x="DEATH_EVENT",data=data,palette=["#DA0858","#3498DB"])
plt.ylabel("Count")
plt.xlabel("Death Event")
plt.show()
```

```python
sns.countplot(x="diabetes",data=data,palette=["#DB0400","#1338BB"])
plt.ylabel("Count")
plt.xlabel("diabetes")
plt.show()
```

```python
pd.crosstab(data.diabetes,data.DEATH_EVENT).plot(kind="bar",figsize=(10,5),color=['#11A5AA','#AA1190'])
plt.title("Distribution of Heart Failure by Diabetes")
plt.xlabel('Diabetes')
plt.xticks(rotation = 0)
plt.legend(["Heart Not Failed", "Heart Failed"])
plt.show()
```

```python
corr_mat = data.corr()
fig, ax = plt.subplots(figsize=(15, 10))
ax = sns.heatmap(corr_mat,annot=True,fmt='.3f')
plt.title("Correlation Matrix",fontsize=30)
plt.show()
```
