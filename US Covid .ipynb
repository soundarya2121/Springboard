```python
df_hi_states = df[df.positive >= 1000000]
df_hi_states
```

```python
plt.figure(figsize=(6, 9))
plt.bar(df_hi_states['state'], df_hi_states['positive'], color='red')
plt.title('Three States With Highest COVID-19 Cases')
plt.xlabel('Top 3  Covid States')
plt.ylabel('Amount of Cases in the Millions')
```

```python
df_CT = df[df.state == 'CT']
df_CT.head()
```

```python
df_CT.sort_values(by='date', inplace=True)
df_CT
```

```python
plt.plot(df_CT['date'], df_CT['positive'], color='red')
plt.title('COVID-19 Positive Cases in Connecticut')
plt.ylabel('Number of Cases in Millions')
plt.xlabel('From 03-15-20 - 04-13-21')
plt.xticks([])
plt.show()
```

```python
plt.plot(df_CT['date'], df_CT['death'], color='black')
plt.title('COVID-19 Deaths in Connecticut')
plt.ylabel('Number of Deaths')
plt.xlabel('From 03-15-20 - 04-13-21')
plt.xticks([])
plt.show()
```

```python
plt.plot(df_CT['date'], df_CT['positive'], color='red', label='Positive')
plt.plot(df_CT['date'], df_CT['death'], color ='black', label='Death')
plt.title('COVID-19 Stats in Connecticut')
plt.ylabel('Number of Cases in Millions')
plt.xlabel('From 03-15-20 - 04-13-21')
plt.legend()
plt.xticks([])
plt.show()
```
